```python
import matplotlib.pyplot as plt

secuencial = 1.145932
concurrente = [1.324407, 0.689606, 0.433663, 0.309603, 0.254769]
concurrente_hilos = [1, 2, 4, 8, 16]
paralelo = [0.283663, 0.287448, 0.284282, 0.283221, 0.283413]
paralelo_hilos = [64, 128, 256, 512, 1024]
plt.plot(concurrente_hilos, concurrente, label='Concurrente', marker='o')
plt.plot(paralelo_hilos, paralelo, label='Paralelo', marker='o')
plt.axhline(secuencial, label='Secuencial', color='red', linestyle='--')
plt.xlabel('Número de hilos')
plt.ylabel('Tiempo (s)')
plt.title('Comparación de tiempos de ejecución')
plt.grid()
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# tiempos promedio de ejecuciones
tiempo_secuencial = secuencial  # promedio de sus resultados de ejecución del algoritmo secuencial

# Datos del CPU multihilado con 1, 2, 4, 8, 16, 32 hilos
hilos_cpu = concurrente_hilos
tiempos_cpu = concurrente  # datos reales de ejecuciones
speedup_cpu = [tiempo_secuencial / tiempo for tiempo in tiempos_cpu]
print(speedup_cpu)

# Datos de la GPU con tamaño de bloque = 64, 128, 256, 512, 1024
hilos_gpu = paralelo_hilos
tiempos_gpu = paralelo  # datos reales de ejecuciones
speedup_gpu = [tiempo_secuencial / tiempo for tiempo in tiempos_gpu]
print(speedup_gpu)

# Ajustar el tamaño de la figura
plt.figure(figsize=(12, 6))  # Ancho de 12 pulgadas y alto de 6 pulgadas

# Primera gráfica: tiempos de ejecución
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primera gráfica
plt.axhline(y=tiempo_secuencial, color='r', linestyle='--', label='Secuencial')
plt.plot(hilos_cpu, tiempos_cpu, marker='o', label='Multihilo CPU')
plt.plot(hilos_gpu, tiempos_gpu, marker='s', label='Multihilo GPU')
plt.title('Gráfica de los tiempos de ejecución')
plt.xlabel('Número de hilos o tamaño del bloque')
plt.ylabel('Tiempo de ejecución promedio (s)')
plt.grid()
plt.legend()

# Segunda gráfica: speedup
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segunda gráfica
plt.plot(hilos_cpu, speedup_cpu, marker='o', label='Speedup CPU')
plt.plot(hilos_gpu, speedup_gpu, marker='s', label='Speedup GPU')  # Corrección del label
plt.xlabel('Número de hilos o tamaño del bloque')
plt.ylabel('Speedup')
plt.title('Gráfica de speedup')
plt.grid()
plt.legend()

plt.show()
```
